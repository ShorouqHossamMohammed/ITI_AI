## using `numpy`

```python
def analytical_pdf (x,mu,sigma):
    return (1/(sigma*np.sqrt(2*np.pi))) * np.exp((-1/2)*((x-mu)/sigma)**2)
```

```python
#analytical pdf 
x = np.linspace(-5,5,100)
y = analytical_pdf (x,0,1)
plt.plot(x,y,color = 'r')

#expremintal pdf

x = np.random.normal(loc = 0,scale = np.sqrt(1),size = 100000)     #np.sqrt(varience)=std
plt.hist(x,density = True , bins = 100);

print("Sample mean: ",x.mean())
print("Sample variance: ",x.var())
```

# Problem 3: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.


```python
rvs = np.random.uniform(low = -1,high=0,size=100000)
lst = []
for i in range(1000):
    lst.append(np.random.choice(rvs,100).mean())
plt.hist(lst,bins=100)
plt.grid(True)
```
