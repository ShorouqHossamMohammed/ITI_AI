## Remove numbers from Company Name

```python
split = after_duplicated_drop['Company Name'].astype(str).str.split('\n')

new_values = []
for val in split:
    val = val[0]
    new_values.append(val)
    
after_duplicated_drop['Company Name']=new_values
after_duplicated_drop['Company Name']
```

## Get values count of 'Salary Estimate' column

```python
after_duplicated_drop['Salary Estimate'].value_counts()
```

## Get the lower and higher limit of the salary in new columns (Hint: use regex, split)

```python
# upper limits
split_salary = after_duplicated_drop['Salary Estimate'].astype(str).str.split('(')
max_salary = []
for val in split_salary:
    val = val[0].split('-')[0].replace("K","").replace("$","")
    max_salary.append(val)

max_salary    
after_duplicated_drop['max_salary']=max_salary
```

```python
# lower limits
split_salary = after_duplicated_drop['Salary Estimate'].astype(str).str.split('(')
min_salary = []
for val in split_salary:
    val = val[0].split('-')[1].replace("K","").replace("$","")
    min_salary.append(val)

min_salary    
after_duplicated_drop['min_salary']=min_salary
after_duplicated_drop
```

## get average salary

```python
average_salary = []
for x,y in zip(after_duplicated_drop['min_salary'],after_duplicated_drop['max_salary']):
    average_salary.append(round(int(x)+int(y)/2))
                
after_duplicated_drop['average_salary']=average_salary
after_duplicated_drop
```

## Extract state code from "location" column as a new column "state"

```python
state_split = after_duplicated_drop['Location'].str.split(",")
state= []
for val in state_split:
    val[0]=val[0].replace(","," ")
    state.append(val[0])

after_duplicated_drop["state"]  = state
after_duplicated_drop
```

## Plot mean salary per Sector using the mean of avg salary column you created earlier

```python
import matplotlib.pyplot as plt
```

```python
mean_salary_per_sector = new_df.groupby('Sector')['average_salary'].mean()

mean_salary_per_sector.plot(kind='barh', color='skyblue')
plt.title('Mean Salary per Sector')
plt.xlabel('Sector')
plt.ylabel('Mean Salary')
plt.show()
```

## Plot count of each occurance of the skills you created earlier

```python
skills_count = new_df['Job Description'].apply(lambda x: [skill for skill in word_list if skill in x.lower()]).explode().value_counts()

skills_count.plot(kind='bar', color='lightgreen')
plt.title('Count of Occurrences of Skills')
plt.xlabel('Skills')
plt.ylabel('Count')
plt.show()
```
