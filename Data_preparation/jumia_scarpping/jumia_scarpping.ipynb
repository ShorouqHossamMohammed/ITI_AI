```python
from selenium import webdriver
import pandas as pd
import requests
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.jumia.com.eg/ar/phones-tablets/")
```

```python
articles = driver.find_elements("xpath","//article[@class='prd _fb col c-prd']")
```

```python
names = []
prices = []
ratings = []

for article in articles:
    try:
        name = article.find_element("xpath",".//h3[@class = 'name']")
        price = article.find_element("xpath",".//div[@class = 'prc']")
        rate = article.find_element("xpath",".//a[@class='core']")
        rate_value = rate.get_attribute("data-dimension27")
        
    except:
        print('error')
       
    if name.text and price.text and rate_value is not None:
        names.append(name.text)
        prices.append(price.text)
        ratings.append(rate_value)
        
```

```python
image_links = [] 
img_path = driver.find_elements("xpath","//article//div//img")

for img in img_path:
    data_src_value = img.get_attribute("data-src")
    print(data_src_value)
    if data_src_value is not None:
        image_links.append(data_src_value)

```

```python
for i in range(len(image_links)):
    try:
        response = requests.get(image_links[i])
        if response.status_code == 200:
            with open("image{}.jpg".format(i),'wb') as M :
                M.write(response.content)
    except:
        print("error")
```

```python
data = {"Name": names,"Price": prices,"Rate": ratings,"Img_path": image_links}
df  = pd.DataFrame(data)
df
```

```python
df.to_csv("data_scrapping.csv")
```
