```python
from selenium import webdriver
import pandas as pd
import requests
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(" https://www.amazon.in/gp/bestsellers/books/")
```

```python
articles = driver.find_elements("xpath","//div[@class ='a-column a-span12 a-text-center _cDEzb_grid-column_2hIsc']")
```

```python
names = []
prices = []

for article in articles:
    try:
        name = article.find_element("xpath",".//div[@class='_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y']")
        price = article.find_element("xpath",".//span[@class = '_cDEzb_p13n-sc-price_3mJ9Z']")
        
    except:
        print('error')
       
    if name.text and price.text  is not None:
        names.append(name.text)
        prices.append(price.text)
        
        
```

```python
data = {"Name": names,"Price": prices}
df  = pd.DataFrame(data)
df
```

```python
df.to_csv("amazon_scrapping.csv")
```
